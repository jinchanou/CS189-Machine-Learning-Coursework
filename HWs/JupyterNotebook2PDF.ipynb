## Notebook to PDF converter

**If running locally:**.
1. Instead of colab, download the notebook by clicking 'save as .ipynb'

2. Open the notebook on your jupyter notebook by going into the folder that contains your notebook (run ```jupyter notebook``` in the terminal. This will allow you to open your notebook in the ```localhost``` environemnt.

3. Then, click download as .tex (to convert it to latex file)

4. Upload the .tex file to overleaf or render it as pdf to get a version that is format-error free.

5. (Alternative 1): option is trying to scale the page to fit on the "print" document. This is changing the Scale -> Custom -> Change from 100 to scale that fits on your page. Now, the cell will fit on each extracted page. However, this does not guarantee that all your each cell elements are format error-free.

6. (Alternative 2): install nbconvert and run `jupyter nbconvert --to pdf PATH_TO_NOTEBOOK`
